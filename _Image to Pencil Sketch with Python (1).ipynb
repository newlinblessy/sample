#### Importing libraries

```python
import cv2
```

### Read the image


```python
from IPython.display import Image
from PIL import Image as PILImage
import numpy as np
```

```python
image = cv2.imread('pug.jpeg')
```

```python
pil_image = PILImage.fromarray(image)
```

```python
display(pil_image)
```

### Grey Scale Image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("New Dog", gray_image)
```

```python
pil_image = PILImage.fromarray(gray_image)
```

```python
display(pil_image)
```

### Invert the new grayscale image

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
```

```python
pil_image = PILImage.fromarray(inverted_image)
display(pil_image)
```

### Blurred Image

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
pil_image = PILImage.fromarray(blurred)
display(pil_image)
```

### Pencil Sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
pil_image = PILImage.fromarray(pencil_sketch)
display(pil_image)
```
